# Text Preprocessing

Preprocessing on textual data are usually : </br>
*Lowercase</br>
*Tokenization</br>
*Removing special characters</br>
*Removing stop words and punctuations</br>
*Stemming

```python
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
stopwords.words("english")
```

```python
import string
string.punctuation
```

```python
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def preprocessing(text):
    text=text.lower()
    text=nltk.word_tokenize(text)
    
    temp=[]
    for i in text:
        if i.isalnum():
            temp.append(i)
    
    text=temp[:]
    temp.clear()
    
    for i in text:
        if i not in stopwords.words("english") and i not in string.punctuation:
            temp.append(ps.stem(i))
            
    return " ".join(temp)
```

```python
#example
preprocessing('Rock-Hard La Varlope, RARE & POWERFUL, HANDSOME JACKPOT, Borderlands 3 (Xbox) dlvr.it/RMTrgF')
```

```python
df['transformed_content']=df['content'].apply(preprocessing)
```

```python
positive_tweet=[]

for tweet in df[df['target']=='Positive']['transformed_content'].tolist():
    for word in tweet.split():
        positive_tweet.append(word)
```

```python
from collections import Counter
positive_tweet_counter=Counter(positive_tweet).most_common(30)
```

```python
positive_df=pd.DataFrame(positive_tweet_counter,columns=['Positive Words','Frequency'])
```

```python
plt.figure(figsize=(10,10))
sns.barplot(x="Positive Words",y="Frequency",data=positive_df)
plt.xticks(rotation='vertical')
plt.show
```
